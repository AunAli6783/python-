```python
import pandas as pd
```

```python
df=pd.read_csv("student_scores.csv")
```

```python
df
```

```python
df['math_score'].mean()
df['english_score'].mean()
df['science_score'].mean()
```

```python
df['sum_Score']=df['math_score']+df['english_score']+df['science_score']
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df.max()
```

```python
df.groupby('gender')['sum_Score'].mean()

```

```python
def grade_ass(score):
  if score>250:
    return 'A'
  elif score>200:
    return 'B'
  else:
    return 'C'

df['grade']=df['sum_Score'].apply(grade_ass)
```

```python
df.to_csv('student_scores_updated.csv')
```
